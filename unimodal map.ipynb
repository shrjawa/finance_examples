```python
#unimodal map of f(x)=27/16 *rx^2 (1-x)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def unimodal_map(r,x):
    return 27*r*(x**2)*(1-x)/16

def cobweb(r,x_0,itern,i):
    f= lambda x: unimodal_map(r,x)
    x=np.linspace(0,1,400)
    plt.figure(i)
    plt.plot(x,f(x))
    plt.plot(x,x)
    
    y=x_0
    for i in range(itern):
        y1=f(y)
        plt.plot([y,y],[y,y1])
        plt.plot([y, y1], [y1, y1])
        y=y1

r=2.9

x_0=0.3
x_1=0.25
itern=1000

cobweb(r,x_0,itern,1)
cobweb(r,x_1,itern,2)
```
