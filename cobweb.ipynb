```python
import numpy as np
import matplotlib.pyplot as plt

def logistic_f(r,x):
    return r*x*(1-x)

def cobweb(r,x_0,itern):
    f= lambda x: logistic_f(r,x)
    x=np.linspace(0,1,400)
    plt.plot(x,f(x))
    plt.plot(x,x)
    
    y=x_0
    for i in range(itern):
        y1=f(y)
        plt.plot([y,y],[y,y1])
        plt.plot([y, y1], [y1, y1])
        y=y1

r=2.9

x_0=0.1

itern=1000

cobweb(r,x_0,itern)
```
