```python
#following example is of lorenz differential equation using runge kutta 4th order numerical method 
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def lorenz_differentials(x, y, z, sigma=10, r=28, b=2.667):
    dx_dt = sigma * (y - x)
    dy_dt = r * x - y - x * z
    dz_dt = x * y - b * z
    return np.array([dx_dt, dy_dt, dz_dt])

def rk4_step(xyz, dt):
    k1 = lorenz_differentials(*xyz)
    k2 = lorenz_differentials(*(xyz + dt/2 * k1))
    k3 = lorenz_differentials(*(xyz + dt/2 * k2))
    k4 = lorenz_differentials(*(xyz + dt * k3))
    return xyz + dt/6 * (k1 + 2*k2 + 2*k3 + k4)

# Simulation parameters
dt = 0.01
num_steps = 5000

# Initial conditions
xyz = np.array([0.0, 1.0, 1.05])
xyzs = np.empty((num_steps + 1, 3))
xyzs[0] = xyz

# Integrate the system
for i in range(num_steps):
    xyzs[i + 1] = rk4_step(xyzs[i], dt)

# Plotting
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.plot(xyzs[:, 0], xyzs[:, 1], xyzs[:, 2], lw=0.5)
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_zlabel("Z Axis")
ax.set_title("Lorenz Attractor with RK4")
plt.show()
```
