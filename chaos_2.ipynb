```python
import matplotlib.pyplot as plt
import math
import numpy as np
def bif(r_min,r_max,step,x_0,itern,itern_1):
    rvals=np.linspace(r_min,r_max,step)
    x=np.zeros(itern)
    for r in rvals:    
        x[0]=x_0
        for i in range(1,itern):
            x[i]=r *np.sin(np.pi * x[i-1])
        plt.plot([r]*itern_1, x[-itern_1:], ',k', alpha=0.25)
        
    plt.xlim(r_min, r_max)
    plt.ylim(0, 1)
    plt.xlabel('r')
    plt.ylabel('x')
    plt.show()
    
r_min=0.1
r_max=1
step=1000
x_0=0.1
itern=1000
itern_1=100
bif(r_min,r_max,step,x_0,itern,itern_1)
```
